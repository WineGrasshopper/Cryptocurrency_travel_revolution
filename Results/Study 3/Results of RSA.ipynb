```python
#    
df['CPT_sq'] = df['CPT'] ** 2
df['PFT_sq'] = df['PFT'] ** 2
df['CPT_PFT'] = df['CPT'] * df['PFT']

#     (   )
X = df[['CPT', 'PFT', 'CPT_sq', 'PFT_sq', 'CPT_PFT', 'MEV', 'DLGroup']]
y = df['CTU']

X = sm.add_constant(X)  #  
model = sm.OLS(y, X).fit()

#  
print(model.summary())

#      
b1 = model.params['CPT']
b2 = model.params['PFT']
b3 = model.params['CPT_sq']  # CPT  
b4 = model.params['PFT_sq']  # PFT  
b5 = model.params['CPT_PFT']  # CPT PFT  

#  (CPT = PFT)
omega1 = b1 + b2  # 
omega2 = b3 + b4 + b5  # 

#  (CPT = -PFT)
omega3 = b1 - b2  # 
omega4 = b3 - b4 + b5  # 

#   
cov_matrix = model.cov_params()

# 1, 2   
omega1_se = np.sqrt(cov_matrix.loc['CPT', 'CPT'] + cov_matrix.loc['PFT', 'PFT'] + 2 * cov_matrix.loc['CPT', 'PFT'])
omega2_se = np.sqrt(cov_matrix.loc['CPT_sq', 'CPT_sq'] + cov_matrix.loc['PFT_sq', 'PFT_sq'] + cov_matrix.loc['CPT_PFT', 'CPT_PFT']
                    + 2 * (cov_matrix.loc['CPT_sq', 'PFT_sq'] + cov_matrix.loc['CPT_sq', 'CPT_PFT'] + cov_matrix.loc['PFT_sq', 'CPT_PFT']))

# 3, 4   
omega3_se = np.sqrt(cov_matrix.loc['CPT', 'CPT'] + cov_matrix.loc['PFT', 'PFT'] - 2 * cov_matrix.loc['CPT', 'PFT'])
omega4_se = np.sqrt(cov_matrix.loc['CPT_sq', 'CPT_sq'] + cov_matrix.loc['PFT_sq', 'PFT_sq'] + cov_matrix.loc['CPT_PFT', 'CPT_PFT']
                    - 2 * (cov_matrix.loc['CPT_sq', 'PFT_sq'] + cov_matrix.loc['CPT_sq', 'CPT_PFT'] - cov_matrix.loc['PFT_sq', 'CPT_PFT']))

# t-  p- 
omega1_t = omega1 / omega1_se
omega2_t = omega2 / omega2_se
omega3_t = omega3 / omega3_se
omega4_t = omega4 / omega4_se

omega1_p = 2 * (1 - norm.cdf(np.abs(omega1_t)))  #  
omega2_p = 2 * (1 - norm.cdf(np.abs(omega2_t)))  #  
omega3_p = 2 * (1 - norm.cdf(np.abs(omega3_t)))  #  
omega4_p = 2 * (1 - norm.cdf(np.abs(omega4_t)))  #  

#  
print(f"Balanced line (CPT = PFT):\n Slope (1) = {omega1} (SE = {omega1_se}, t = {omega1_t}, p = {omega1_p})\n Curvature (2) = {omega2} (SE = {omega2_se}, t = {omega2_t}, p = {omega2_p})")
print(f"Imbalanced line (CPT = -PFT):\n Slope (3) = {omega3} (SE = {omega3_se}, t = {omega3_t}, p = {omega3_p})\n Curvature (4) = {omega4} (SE = {omega4_se}, t = {omega4_t}, p = {omega4_p})")
```
